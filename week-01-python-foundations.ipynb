## 2. `return` 决定函数交回什么

`return` 应在结果准备好后执行。若把它放进循环，函数通常只检查第一个元素就提前结束。

需求：`even_mean(numbers)` 返回所有偶数的算术平均数；若没有偶数，返回 `None`。

先自己写，再对照下面的参考实现。注意平均数使用 `/`，不是向下取整的 `//`。


```python
def even_mean(numbers):
    evens = []
    for number in numbers:
        if number % 2 == 0:
            evens.append(number)

    if not evens:
        return None
    return sum(evens) / len(evens)


print(even_mean([1, 2, 3, 4, 8]))
```

### 自动测试：这些断言全部通过才算完成


```python
assert even_mean([1, 2, 3, 4, 8]) == 14 / 3
assert even_mean([2]) == 2.0
assert even_mean([-4, -1, 2]) == -1.0
assert even_mean([1, 3, 5]) is None
assert even_mean([]) is None
print("even_mean：5/5 测试通过 ✅")
```

## 3. 字典：给统计结果命名

机器学习实验不能只说“效果不错”，要记录样本数、偶数数目和均值。字典让这些字段可读。


```python
def summarize_even_numbers(numbers):
    evens = [number for number in numbers if number % 2 == 0]
    return {
        "total": len(numbers),
        "even_count": len(evens),
        "even_mean": sum(evens) / len(evens) if evens else None,
    }


summary = summarize_even_numbers([1, 2, 4, 7, 8])
summary
```

```python
assert summary == {"total": 5, "even_count": 3, "even_mean": 14 / 3}
assert summarize_even_numbers([])["even_count"] == 0
print("统计函数测试通过 ✅")
```

## 常见错误

- `a = (x)` 不是“装着 x 的列表”；单元素元组应写 `(x,)`，列表应写 `[x]`。
- `x = y` 会覆盖旧值；收集多个值应使用 `result.append(y)`。
- 在 `else` 中 `return None` 会让函数遇到第一个奇数就结束。
- `sum(values) // len(values)` 会丢失小数；平均数通常用 `/`。
- 没有偶数时直接除以 `len([])` 会触发 `ZeroDivisionError`。


## 扩展挑战

不修改上面的函数，另写 `positive_even_mean(numbers)`：只计算**正偶数**平均数。  
最少补三个测试：普通输入、没有正偶数、同时含负数和零。


```python
def positive_even_mean(numbers):
    values = [number for number in numbers if number > 0 and number % 2 == 0]
    return sum(values) / len(values) if values else None


assert positive_even_mean([-4, 0, 2, 6]) == 4.0
assert positive_even_mean([-2, 0, 3]) is None
assert positive_even_mean([10]) == 10.0
print("扩展挑战参考实现测试通过 ✅")
```
